### Styling Lines and Markers

The `plt.plot` function supports many arguments for styling lines and markers:

- `color` or `c`: Set the color of the line ([supported colors](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` or `ls`: Choose between a solid or dashed line
- `linewidth` or `lw`: Set the width of a line
- `markersize` or `ms`: Set the size of markers
- `markeredgecolor` or `mec`: Set the edge color for markers
- `markeredgewidth` or `mew`: Set the edge width for markers
- `markerfacecolor` or `mfc`: Set the fill color for markers
- `alpha`: Opacity of the plot


Check out the documentation for `plt.plot` to learn more: [https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) .

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

the fmt argument provides a shorthand for specifying the marker shape,line styple and line color . it can be provided as the third argument to plt.plot

fmt=`[marker][line][color]`

```python
plt.plot(years,apples,'s--b')
plt.plot(years,oranges,'o-r')

plt.xlabel('year')
plt.ylabel('yield (tons per hectare)')

plt.title("Crop yields in kanato")
plt.legend(['Apples','oranges']);
```

```python
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

we can use scatter plot to visualize how sepal length and width vary using the scatter pllot function from seaborm (imported as sns)

seaborn has inbuilt support for pandas data frames. Instead of passing each column as a series, you can also pass column names and use the data argument to pass the data frame

```python
plt.title('sepal dimensions')
sns.scatterplot(x='sepal_length',
                y='sepal_width',
                hue='species',
                s=100,
                data=flowers_df);
```

# HIstogram
it represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of the observations that fall in each bin

```python
# load data into a pandas dataframe
flowers_df= sns.load_dataset("iris")
flowers_df.sepal_width
```

```python
flowers_df.describe()
```

```python
plt.title("distribution of sepal width")
plt.hist(flowers_df.sepal_width);

# each interval is called bin
# interval is mathematical terms and binis statistical term
```

we can see that values of sepal width fall in the rage of 2.9 to 3.1 are around 35 values 

## controlling the size and number of bins


```python
#specifying the number of bins
plt.hist(flowers_df.sepal_width,bins=5);
```

```python
import numpy as np
# import matplotlib.pyplot as plt
# np.arange(2,5,0.25)
# Specifying the boundaries of each bin
sns.histplot(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25),kde=True);
```

```python
# Bins of unequal sizes
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]);
```

## multiple historgram
similar to line chart we can draw multile histogram in a single chart. we can reduce each histogram opacity so tht one histogram bars dont hide the others

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.7, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));

plt.legend(['Setosa', 'versicolor']);
```

we can also stack multiple histogram on top of another

```python
plt.title('Distribution of sepal width')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True);
```

## box plots with averages
lets look at the another sample dataset included with seaborn called tips . the dataset contains information about the sex, time of the day, total bill and tip amount for customers visiting a resturant over a week

```python
tips_df=sns.load_dataset("tips")
```

```python
tips_df
```

```python
bill_avg_df=tips_df.groupby('day')[['total_bill']].mean()
bill_avg_df
```

```python
plt.bar(bill_avg_df.index,bill_avg_df.total_bill)
```

##  we might want to draw a bar chart to visualise how the average bill amount varies across different days of the week. one way to do this would be to compute the day wise averages and then use plt.bar 

## however since this is a very common use case, the seaborn library provides a barplot function which can automatically compute averages

```python
tips_df
```

```python
sns.barplot(x='day',y='total_bill',data=tips_df);
```

### the lines cutting each bar repressents the amount of variation in the values. for instance, it seems like the variation in the total bill was quite high on fridays, and lowero n suturdays.

### we can also specify a hue argument to compare bar plots side by side based on a third features eg sex

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df);
```

### you can make bars horizontal simply by switching the axes.

```python
sns.barplot(x='total_bill',y='day',hue='sex',data=tips_df);
```
